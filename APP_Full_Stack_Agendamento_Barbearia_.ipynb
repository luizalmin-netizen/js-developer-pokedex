```java
import java.time.LocalDateTime;

public class Agendamento {

    private Long id;
    private Cliente cliente;
    private Barbeiro barbeiro;
    private Servico servico;
    private LocalDateTime dataHoraAgendamento;
    private String status; // Ex: PENDENTE, CONFIRMADO, CANCELADO, CONCLUIDO
    private String observacoes;

    // Construtor vazio
    public Agendamento() {
    }

    // Construtor com todos os campos
    public Agendamento(Long id, Cliente cliente, Barbeiro barbeiro, Servico servico, LocalDateTime dataHoraAgendamento, String status, String observacoes) {
        this.id = id;
        this.cliente = cliente;
        this.barbeiro = barbeiro;
        this.servico = servico;
        this.dataHoraAgendamento = dataHoraAgendamento;
        this.status = status;
        this.observacoes = observacoes;
    }

    // Getters e Setters
    public Long getId() {
        return id;
    }

    public void setId(Long id) {
        this.id = id;
    {}

    public Cliente getCliente() {
        return cliente;
    }

    public void setCliente(Cliente cliente) {
        this.cliente = cliente;
    }

    public Barbeiro getBarbeiro() {
        return barbeiro;
    }

    public void setBarbeiro(Barbeiro barbeiro) {
        this.barbeiro = barbeiro;
    }

    public Servico getServico() {
        return servico;
    }

    public void setServico(Servico servico) {
        this.servico = servico;
    }

    public LocalDateTime getDataHoraAgendamento() {
        return dataHoraAgendamento;
    }

    public void setDataHoraAgendamento(LocalDateTime dataHoraAgendamento) {
        this.dataHoraAgendamento = dataHoraAgendamento;
    }

    public String getStatus() {
        return status;
    }

    public void setStatus(String status) {
        this.status = status;
    }

    public String getObservacoes() {
        return observacoes;
    }

    public void setObservacoes(String observacoes) {
        this.observacoes = observacoes;
    }

    // Opcional: toString, equals, hashCode
    @Override
    public String toString() {
        return "Agendamento{" +
               "id=" + id +
               ", cliente=" + (cliente != null ? cliente.getNome() : "N/A") +
               ", barbeiro=" + (barbeiro != null ? barbeiro.getNome() : "N/A") +
               ", servico=" + (servico != null ? servico.getNome() : "N/A") +
               ", dataHoraAgendamento=" + dataHoraAgendamento +
               ", status='" + status + '\'' +
               ", observacoes='" + observacoes + '\'' +
               '}';
    }
}

// Nota: Você precisaria definir as classes Cliente, Barbeiro e Servico separadamente.

```java
public class Cliente {

    private Long id;
    private String nome;
    private String telefone;
    private String email;

    public Cliente() {
    }

    public Cliente(Long id, String nome, String telefone, String email) {
        this.id = id;
        this.nome = nome;
        this.telefone = telefone;
        this.email = email;
    }

    public Long getId() {
        return id;
    }

    public void setId(Long id) {
        this.id = id;
    }

    public String getNome() {
        return nome;
    }

    public void setNome(String nome) {
        this.nome = nome;
    }

    public String getTelefone() {
        return telefone;
    }

    public void setTelefone(String telefone) {
        this.telefone = telefone;
    }

    public String getEmail() {
        return email;
    }

    public void setEmail(String email) {
        this.email = email;
    }

    @Override
    public String toString() {
        return "Cliente{" +
               "id=" + id +
               ", nome='" + nome + '\'' +
               ", telefone='" + telefone + '\'' +
               ", email='" + email + '\'' +
               '}';
    }
}


```java
public class Barbeiro {

    private Long id;
    private String nome;
    private String especialidade;
    private String telefone;
    private String email;

    public Barbeiro() {
    }

    public Barbeiro(Long id, String nome, String especialidade, String telefone, String email) {
        this.id = id;
        this.nome = nome;
        this.especialidade = especialidade;
        this.telefone = telefone;
        this.email = email;
    }

    public Long getId() {
        return id;
    }

    public void setId(Long id) {
        this.id = id;
    }

    public String getNome() {
        return nome;
    }

    public void setNome(String nome) {
        this.nome = nome;
    }

    public String getEspecialidade() {
        return especialidade;
    }

    public void setEspecialidade(String especialidade) {
        this.especialidade = especialidade;
    }

    public String getTelefone() {
        return telefone;
    }

    public void setTelefone(String telefone) {
        this.telefone = telefone;
    }

    public String getEmail() {
        return email;
    }

    public void setEmail(String email) {
        this.email = email;
    }

    @Override
    public String toString() {
        return "Barbeiro{" +
               "id=" + id +
               ", nome='" + nome + '\'' +
               ", especialidade='" + especialidade + '\'' +
               ", telefone='" + telefone + '\'' +
               ", email='" + email + '\'' +
               '}';
    }
}


```java
public class Servico {

    private Long id;
    private String nome;
    private String descricao;
    private Double preco;
    private Integer duracaoEstimadaEmMinutos;

    public Servico() {
    }

    public Servico(Long id, String nome, String descricao, Double preco, Integer duracaoEstimadaEmMinutos) {
        this.id = id;
        this.nome = nome;
        this.descricao = descricao;
        this.preco = preco;
        this.duracaoEstimadaEmMinutos = duracaoEstimadaEmMinutos;
    }

    public Long getId() {
        return id;
    }

    public void setId(Long id) {
        this.id = id;
    }

    public String getNome() {
        return nome;
    }

    public void setNome(String nome) {
        this.nome = nome;
    }

    public String getDescricao() {
        return descricao;
    }

    public void setDescricao(String descricao) {
        this.descricao = descricao;
    }

    public Double getPreco() {
        return preco;
    }

    public void setPreco(Double preco) {
        this.preco = preco;
    }

    public Integer getDuracaoEstimadaEmMinutos() {
        return duracaoEstimadaEmMinutos;
    }

    public void setDuracaoEstimadaEmMinutos(Integer duracaoEstimadaEmMinutos) {
        this.duracaoEstimadaEmMinutos = duracaoEstimadaEmMinutos;
    }

    @Override
    public String toString() {
        return "Servico{" +
               "id=" + id +
               ", nome='" + nome + '\'' +
               ", descricao='" + descricao + '\'' +
               ", preco=" + preco +
               ", duracaoEstimadaEmMinutos=" + duracaoEstimadaEmMinutos +
               '}';
    }
}


```java
import org.springframework.data.jpa.repository.JpaRepository;
import org.springframework.stereotype.Repository;

@Repository
public interface ClienteRepository extends JpaRepository<Cliente, Long> {
    // Métodos personalizados podem ser adicionados aqui
    Cliente findByEmail(String email);
}


```java
import org.springframework.data.jpa.repository.JpaRepository;
import org.springframework.stereotype.Repository;

@Repository
public interface BarbeiroRepository extends JpaRepository<Barbeiro, Long> {
    // Métodos personalizados podem ser adicionados aqui
    Barbeiro findByNome(String nome);
}


```java
import org.springframework.data.jpa.repository.JpaRepository;
import org.springframework.stereotype.Repository;

@Repository
public interface ServicoRepository extends JpaRepository<Servico, Long> {
    // Métodos personalizados podem ser adicionados aqui
    Servico findByNome(String nome);
}


```java
import org.springframework.data.jpa.repository.JpaRepository;
import org.springframework.stereotype.Repository;
import java.time.LocalDateTime;
import java.util.List;

@Repository
public interface AgendamentoRepository extends JpaRepository<Agendamento, Long> {
    // Métodos personalizados para consultas de agendamentos
    List<Agendamento> findByCliente(Cliente cliente);
    List<Agendamento> findByBarbeiro(Barbeiro barbeiro);
    List<Agendamento> findByDataHoraAgendamentoBetween(LocalDateTime start, LocalDateTime end);
    List<Agendamento> findByBarbeiroAndDataHoraAgendamentoBetween(Barbeiro barbeiro, LocalDateTime start, LocalDateTime end);
    List<Agendamento> findByStatus(String status);
}


```java
import org.springframework.beans.factory.annotation.Autowired;
import org.springframework.stereotype.Service;
import java.time.LocalDateTime;
import java.time.format.DateTimeFormatter;
import java.util.List;
import java.util.Optional;

// Supondo a existência dessas classes de exceção ou crie-as se necessário
class ResourceNotFoundException extends RuntimeException {
    public ResourceNotFoundException(String message) {
        super(message);
    }
}

class AgendamentoConflictException extends RuntimeException {
    public AgendamentoConflictException(String message) {
        super(message);
    }
}

@Service
public class AgendamentoService {

    private final AgendamentoRepository agendamentoRepository;
    private final ClienteRepository clienteRepository;
    private final BarbeiroRepository barbeiroRepository;
    private final ServicoRepository servicoRepository;

    @Autowired
    public AgendamentoService(AgendamentoRepository agendamentoRepository,
                              ClienteRepository clienteRepository,
                              BarbeiroRepository barbeiroRepository,
                              ServicoRepository servicoRepository) {
        this.agendamentoRepository = agendamentoRepository;
        this.clienteRepository = clienteRepository;
        this.barbeiroRepository = barbeiroRepository;
        this.servicoRepository = servicoRepository;
    }

    public List<Agendamento> buscarTodosAgendamentos() {
        return agendamentoRepository.findAll();
    }

    public Optional<Agendamento> buscarAgendamentoPorId(Long id) {
        return agendamentoRepository.findById(id);
    }

    public Agendamento salvarAgendamento(Agendamento agendamento) {
        // Adicionar validações antes de salvar, se necessário
        return agendamentoRepository.save(agendamento);
    }

    public void deletarAgendamento(Long id) {
        if (!agendamentoRepository.existsById(id)) {
            throw new ResourceNotFoundException("Agendamento não encontrado com o ID: " + id);
        }
        agendamentoRepository.deleteById(id);
    }

    // --- Métodos de Lógica de Negócio --- //

    public Agendamento agendarNovoServico(Long clienteId, Long barbeiroId, Long servicoId, LocalDateTime dataHoraAgendamento) {
        Cliente cliente = clienteRepository.findById(clienteId)
                .orElseThrow(() -> new ResourceNotFoundException("Cliente não encontrado com o ID: " + clienteId));
        Barbeiro barbeiro = barbeiroRepository.findById(barbeiroId)
                .orElseThrow(() -> new ResourceNotFoundException("Barbeiro não encontrado com o ID: " + barbeiroId));
        Servico servico = servicoRepository.findById(servicoId)
                .orElseThrow(() -> new ResourceNotFoundException("Serviço não encontrado com o ID: " + servicoId));

        // 1. Calcular o horário de início e fim do agendamento proposto
        LocalDateTime proposedStart = dataHoraAgendamento;
        LocalDateTime proposedEnd = proposedStart.plusMinutes(servico.getDuracaoEstimadaEmMinutos());

        // 2. Definir uma janela de busca para agendamentos existentes que possam conflitar
        //    Para ser abrangente, considera-se que um agendamento pode começar antes do proposto
        //    e terminar dentro dele, ou começar dentro do proposto e terminar depois.
        //    Assumimos uma duração máxima de serviço para a janela de busca, por exemplo, 2 horas (120 minutos).
        //    Isso pode ser refinado para ser dinâmico com base nos serviços cadastrados.
        final int MAX_POSSIBLE_SERVICE_DURATION_FOR_CHECK = 120; // Em minutos, ajustar conforme a lógica de negócio

        LocalDateTime searchWindowStart = proposedStart.minusMinutes(MAX_POSSIBLE_SERVICE_DURATION_FOR_CHECK);
        LocalDateTime searchWindowEnd = proposedEnd;

        // 3. Buscar agendamentos potencialmente conflitantes para o barbeiro dentro desta janela
        List<Agendamento> existingAgendamentos = agendamentoRepository.findByBarbeiroAndDataHoraAgendamentoBetween(
            barbeiro, searchWindowStart, searchWindowEnd
        );

        // 4. Iterar e verificar se há conflito de horário
        for (Agendamento existingAgendamento : existingAgendamentos) {
            // Ignora agendamentos que já foram cancelados ou concluídos, se houver necessidade
            if (existingAgendamento.getStatus().equals("CANCELADO") || existingAgendamento.getStatus().equals("CONCLUIDO")) {
                continue;
            }
            
            // Garante que o serviço do agendamento existente não é nulo
            if (existingAgendamento.getServico() == null) {
                // Pode-se adicionar log ou lançar uma exceção de inconsistência de dados aqui
                continue;
            }

            LocalDateTime existingStart = existingAgendamento.getDataHoraAgendamento();
            LocalDateTime existingEnd = existingStart.plusMinutes(existingAgendamento.getServico().getDuracaoEstimadaEmMinutos());

            // Verifica condição de sobreposição:
            // O novo agendamento começa antes do agendamento existente terminar E
            // O novo agendamento termina depois do agendamento existente começar
            if (proposedStart.isBefore(existingEnd) && proposedEnd.isAfter(existingStart)) {
                throw new AgendamentoConflictException(
                    "Barbeiro " + barbeiro.getNome() + " não disponível para agendamento entre " +
                    proposedStart.format(DateTimeFormatter.ofPattern("dd/MM/yyyy HH:mm")) + " e " +
                    proposedEnd.format(DateTimeFormatter.ofPattern("dd/MM/yyyy HH:mm")) +
                    ". Conflito com agendamento existente às " +
                    existingStart.format(DateTimeFormatter.ofPattern("dd/MM/yyyy HH:mm")) + "."
                );
            }
        }

        Agendamento novoAgendamento = new Agendamento();
        novoAgendamento.setCliente(cliente);
        novoAgendamento.setBarbeiro(barbeiro);
        novoAgendamento.setServico(servico);
        novoAgendamento.setDataHoraAgendamento(dataHoraAgendamento);
        novoAgendamento.setStatus("PENDENTE"); // Status inicial
        novoAgendamento.setObservacoes("Agendamento criado via sistema");

        return agendamentoRepository.save(novoAgendamento);
    }

    public Agendamento confirmarAgendamento(Long agendamentoId) {
        Agendamento agendamento = agendamentoRepository.findById(agendamentoId)
                .orElseThrow(() -> new ResourceNotFoundException("Agendamento não encontrado com o ID: " + agendamentoId));
        agendamento.setStatus("CONFIRMADO");
        return agendamentoRepository.save(agendamento);
    }

    public Agendamento cancelarAgendamento(Long agendamentoId, String motivoCancelamento) {
        Agendamento agendamento = agendamentoRepository.findById(agendamentoId)
                .orElseThrow(() -> new ResourceNotFoundException("Agendamento não encontrado com o ID: " + agendamentoId));
        agendamento.setStatus("CANCELADO");
        agendamento.setObservacoes(agendamento.getObservacoes() + "\nCancelado: " + motivoCancelamento);
        return agendamentoRepository.save(agendamento);
    }

    public List<Agendamento> buscarAgendamentosPorCliente(Long clienteId) {
        Cliente cliente = clienteRepository.findById(clienteId)
                .orElseThrow(() -> new ResourceNotFoundException("Cliente não encontrado com o ID: " + clienteId));
        return agendamentoRepository.findByCliente(cliente);
    }

    public List<Agendamento> buscarAgendamentosPorBarbeiro(Long barbeiroId) {
        Barbeiro barbeiro = barbeiroRepository.findById(barbeiroId)
                .orElseThrow(() -> new ResourceNotFoundException("Barbeiro não encontrado com o ID: " + barbeiroId));
        return agendamentoRepository.findByBarbeiro(barbeiro);
    }

    public List<Agendamento> buscarAgendamentosPorPeriodo(LocalDateTime inicio, LocalDateTime fim) {
        return agendamentoRepository.findByDataHoraAgendamentoBetween(inicio, fim);
    }
}


```java
import org.junit.jupiter.api.BeforeEach;
import org.junit.jupiter.api.Test;
import org.junit.jupiter.api.extension.ExtendWith;
import org.mockito.InjectMocks;
import org.mockito.Mock;
import org.mockito.junit.jupiter.MockitoExtension;

import java.time.LocalDateTime;
import java.util.Collections;
import java.util.Optional;
import java.util.Arrays;

import static org.junit.jupiter.api.Assertions.*;
import static org.mockito.Mockito.*;

@ExtendWith(MockitoExtension.class)
public class AgendamentoServiceTest {

    @Mock
    private AgendamentoRepository agendamentoRepository;
    @Mock
    private ClienteRepository clienteRepository;
    @Mock
    private BarbeiroRepository barbeiroRepository;
    @Mock
    private ServicoRepository servicoRepository;

    @InjectMocks
    private AgendamentoService agendamentoService;

    private Cliente cliente;
    private Barbeiro barbeiro;
    private Servico servico;
    private LocalDateTime dataHoraAgendamento;

    @BeforeEach
    void setUp() {
        cliente = new Cliente(1L, "João", "11987654321", "joao@email.com");
        barbeiro = new Barbeiro(1L, "Carlos", "Corte Masculino", "11912345678", "carlos@email.com");
        servico = new Servico(1L, "Corte de Cabelo", "Corte padrão", 50.0, 30);
        dataHoraAgendamento = LocalDateTime.of(2024, 7, 10, 10, 0);

        when(clienteRepository.findById(1L)).thenReturn(Optional.of(cliente));
        when(barbeiroRepository.findById(1L)).thenReturn(Optional.of(barbeiro));
        when(servicoRepository.findById(1L)).thenReturn(Optional.of(servico));
    }

    @Test
    void agendarNovoServico_sucesso() {
        // Configura o repositório para retornar uma lista vazia, indicando que não há conflitos
        when(agendamentoRepository.findByBarbeiroAndDataHoraAgendamentoBetween(any(Barbeiro.class), any(LocalDateTime.class), any(LocalDateTime.class)))
                .thenReturn(Collections.emptyList());

        // Configura o repositório para retornar o agendamento salvo
        when(agendamentoRepository.save(any(Agendamento.class)))
                .thenAnswer(invocation -> {
                    Agendamento agendamento = invocation.getArgument(0);
                    agendamento.setId(1L); // Simula o ID sendo gerado pelo banco
                    return agendamento;
                });

        Agendamento novoAgendamento = agendamentoService.agendarNovoServico(
                cliente.getId(), barbeiro.getId(), servico.getId(), dataHoraAgendamento);

        assertNotNull(novoAgendamento);
        assertEquals(1L, novoAgendamento.getId());
        assertEquals(cliente, novoAgendamento.getCliente());
        assertEquals(barbeiro, novoAgendamento.getBarbeiro());
        assertEquals(servico, novoAgendamento.getServico());
        assertEquals(dataHoraAgendamento, novoAgendamento.getDataHoraAgendamento());
        assertEquals("PENDENTE", novoAgendamento.getStatus());

        verify(agendamentoRepository, times(1)).findByBarbeiroAndDataHoraAgendamentoBetween(any(), any(), any());
        verify(agendamentoRepository, times(1)).save(any(Agendamento.class));
    }

    @Test
    void agendarNovoServico_conflitoDeHorario() {
        // Cria um agendamento existente que entra em conflito
        Servico servicoExistente = new Servico(2L, "Barba", "Aparar barba", 25.0, 20);
        Agendamento agendamentoConflitante = new Agendamento(
                2L, cliente, barbeiro, servicoExistente, LocalDateTime.of(2024, 7, 10, 10, 15), "CONFIRMADO", "");

        // Configura o repositório para retornar o agendamento conflitante
        when(agendamentoRepository.findByBarbeiroAndDataHoraAgendamentoBetween(any(Barbeiro.class), any(LocalDateTime.class), any(LocalDateTime.class)))
                .thenReturn(Arrays.asList(agendamentoConflitante));

        // Espera que uma AgendamentoConflictException seja lançada
        AgendamentoConflictException exception = assertThrows(AgendamentoConflictException.class, () -> {
            agendamentoService.agendarNovoServico(
                    cliente.getId(), barbeiro.getId(), servico.getId(), dataHoraAgendamento);
        });

        assertTrue(exception.getMessage().contains("não disponível"));
        verify(agendamentoRepository, times(1)).findByBarbeiroAndDataHoraAgendamentoBetween(any(), any(), any());
        verify(agendamentoRepository, never()).save(any(Agendamento.class)); // Não deve salvar em caso de conflito
    }

    @Test
    void agendarNovoServico_clienteNaoEncontrado() {
        when(clienteRepository.findById(anyLong())).thenReturn(Optional.empty());

        assertThrows(ResourceNotFoundException.class, () -> {
            agendamentoService.agendarNovoServico(99L, barbeiro.getId(), servico.getId(), dataHoraAgendamento);
        });
        verify(agendamentoRepository, never()).save(any(Agendamento.class));
    }

    @Test
    void agendarNovoServico_barbeiroNaoEncontrado() {
        when(barbeiroRepository.findById(anyLong())).thenReturn(Optional.empty());

        assertThrows(ResourceNotFoundException.class, () -> {
            agendamentoService.agendarNovoServico(cliente.getId(), 99L, servico.getId(), dataHoraAgendamento);
        });
        verify(agendamentoRepository, never()).save(any(Agendamento.class));
    }

    @Test
    void agendarNovoServico_servicoNaoEncontrado() {
        when(servicoRepository.findById(anyLong())).thenReturn(Optional.empty());

        assertThrows(ResourceNotFoundException.class, () -> {
            agendamentoService.agendarNovoServico(cliente.getId(), barbeiro.getId(), 99L, dataHoraAgendamento);
        });
        verify(agendamentoRepository, never()).save(any(Agendamento.class));
    }

    @Test
    void agendarNovoServico_agendamentoConflitanteCanceladoNaoGeraErro() {
        // Cria um agendamento existente cancelado que se sobreporia
        Servico servicoExistente = new Servico(2L, "Manicure", "", 30.0, 45);
        Agendamento agendamentoCancelado = new Agendamento(
                2L, cliente, barbeiro, servicoExistente, LocalDateTime.of(2024, 7, 10, 9, 45), "CANCELADO", "");

        // Retorna o agendamento cancelado na busca
        when(agendamentoRepository.findByBarbeiroAndDataHoraAgendamentoBetween(any(Barbeiro.class), any(LocalDateTime.class), any(LocalDateTime.class)))
                .thenReturn(Arrays.asList(agendamentoCancelado));

        // Mock para o save, como no teste de sucesso
        when(agendamentoRepository.save(any(Agendamento.class)))
                .thenAnswer(invocation -> {
                    Agendamento agendamento = invocation.getArgument(0);
                    agendamento.setId(1L);
                    return agendamento;
                });

        // O agendamento deve ser criado com sucesso, pois o agendamento conflitante está cancelado
        assertDoesNotThrow(() -> agendamentoService.agendarNovoServico(
                cliente.getId(), barbeiro.getId(), servico.getId(), dataHoraAgendamento));

        verify(agendamentoRepository, times(1)).findByBarbeiroAndDataHoraAgendamentoBetween(any(), any(), any());
        verify(agendamentoRepository, times(1)).save(any(Agendamento.class));
    }
}


```java
import java.time.LocalDateTime;

// Este é um Data Transfer Object (DTO) para receber os dados de criação de um agendamento via API.
// Ajuda a desacoplar a entrada da API do modelo de domínio Agendamento.
public class AgendamentoRequestDTO {
    private Long clienteId;
    private Long barbeiroId;
    private Long servicoId;
    private LocalDateTime dataHoraAgendamento;
    private String observacoes;

    public AgendamentoRequestDTO() {
    }

    public AgendamentoRequestDTO(Long clienteId, Long barbeiroId, Long servicoId, LocalDateTime dataHoraAgendamento, String observacoes) {
        this.clienteId = clienteId;
        this.barbeiroId = barbeiroId;
        this.servicoId = servicoId;
        this.dataHoraAgendamento = dataHoraAgendamento;
        this.observacoes = observacoes;
    }

    // Getters e Setters
    public Long getClienteId() {
        return clienteId;
    }

    public void setClienteId(Long clienteId) {
        this.clienteId = clienteId;
    }

    public Long getBarbeiroId() {
        return barbeiroId;
    }

    public void setBarbeiroId(Long barbeiroId) {
        this.barbeiroId = barbeiroId;
    }

    public Long getServicoId() {
        return servicoId;
    }

    public void setServicoId(Long servicoId) {
        this.servicoId = servicoId;
    }

    public LocalDateTime getDataHoraAgendamento() {
        return dataHoraAgendamento;
    }

    public void setDataHoraAgendamento(LocalDateTime dataHoraAgendamento) {
        this.dataHoraAgendamento = dataHoraAgendamento;
    }

    public String getObservacoes() {
        return observacoes;
    }

    public void setObservacoes(String observacoes) {
        this.observacoes = observacoes;
    }
}


```java
import org.springframework.beans.factory.annotation.Autowired;
import org.springframework.http.HttpStatus;
import org.springframework.http.ResponseEntity;
import org.springframework.web.bind.annotation.*;

import java.util.List;

@RestController
@RequestMapping("/api/agendamentos")
public class AgendamentoController {

    private final AgendamentoService agendamentoService;

    @Autowired
    public AgendamentoController(AgendamentoService agendamentoService) {
        this.agendamentoService = agendamentoService;
    }

    @GetMapping
    public ResponseEntity<List<Agendamento>> getAllAgendamentos() {
        List<Agendamento> agendamentos = agendamentoService.buscarTodosAgendamentos();
        return ResponseEntity.ok(agendamentos);
    }

    @GetMapping("/{id}")
    public ResponseEntity<Agendamento> getAgendamentoById(@PathVariable Long id) {
        return agendamentoService.buscarAgendamentoPorId(id)
                .map(ResponseEntity::ok)
                .orElse(ResponseEntity.notFound().build());
    }

    @PostMapping
    public ResponseEntity<Agendamento> createAgendamento(@RequestBody AgendamentoRequestDTO requestDTO) {
        try {
            Agendamento novoAgendamento = agendamentoService.agendarNovoServico(
                    requestDTO.getClienteId(),
                    requestDTO.getBarbeiroId(),
                    requestDTO.getServicoId(),
                    requestDTO.getDataHoraAgendamento()
            );
            // Se houver observações no DTO, pode-se adicioná-las aqui antes de salvar novamente ou passar para o serviço
            if (requestDTO.getObservacoes() != null && !requestDTO.getObservacoes().isEmpty()) {
                 novoAgendamento.setObservacoes(requestDTO.getObservacoes());
                 agendamentoService.salvarAgendamento(novoAgendamento); // Salva as observações
            }
            return new ResponseEntity<>(novoAgendamento, HttpStatus.CREATED);
        } catch (ResourceNotFoundException e) {
            return new ResponseEntity(e.getMessage(), HttpStatus.NOT_FOUND);
        } catch (AgendamentoConflictException e) {
            return new ResponseEntity(e.getMessage(), HttpStatus.CONFLICT);
        } catch (Exception e) {
            return new ResponseEntity("Erro interno do servidor: " + e.getMessage(), HttpStatus.INTERNAL_SERVER_ERROR);
        }
    }

    @PutMapping("/{id}")
    public ResponseEntity<Agendamento> updateAgendamento(@PathVariable Long id, @RequestBody Agendamento agendamentoDetails) {
        return agendamentoService.buscarAgendamentoPorId(id)
                .map(agendamentoExistente -> {
                    // Atualizar os campos necessários
                    agendamentoExistente.setCliente(agendamentoDetails.getCliente());
                    agendamentoExistente.setBarbeiro(agendamentoDetails.getBarbeiro());
                    agendamentoExistente.setServico(agendamentoDetails.getServico());
                    agendamentoExistente.setDataHoraAgendamento(agendamentoDetails.getDataHoraAgendamento());
                    agendamentoExistente.setStatus(agendamentoDetails.getStatus());
                    agendamentoExistente.setObservacoes(agendamentoDetails.getObservacoes());
                    Agendamento agendamentoAtualizado = agendamentoService.salvarAgendamento(agendamentoExistente);
                    return ResponseEntity.ok(agendamentoAtualizado);
                }).orElse(ResponseEntity.notFound().build());
    }

    @DeleteMapping("/{id}")
    public ResponseEntity<Void> deleteAgendamento(@PathVariable Long id) {
        try {
            agendamentoService.deletarAgendamento(id);
            return ResponseEntity.noContent().build();
        } catch (ResourceNotFoundException e) {
            return ResponseEntity.notFound().build();
        }
    }

    @PostMapping("/{id}/confirmar")
    public ResponseEntity<Agendamento> confirmarAgendamento(@PathVariable Long id) {
        try {
            Agendamento agendamentoConfirmado = agendamentoService.confirmarAgendamento(id);
            return ResponseEntity.ok(agendamentoConfirmado);
        } catch (ResourceNotFoundException e) {
            return ResponseEntity.notFound().build();
        }
    }

    @PostMapping("/{id}/cancelar")
    public ResponseEntity<Agendamento> cancelarAgendamento(@PathVariable Long id, @RequestBody(required = false) String motivoCancelamento) {
        try {
            // O motivo de cancelamento pode ser opcional ou vir de um DTO mais complexo
            Agendamento agendamentoCancelado = agendamentoService.cancelarAgendamento(id, motivoCancelamento != null ? motivoCancelamento : "Cancelado pelo usuário");
            return ResponseEntity.ok(agendamentoCancelado);
        } catch (ResourceNotFoundException e) {
            return ResponseEntity.notFound().build();
        }
    }

    // Exemplo de endpoint para buscar agendamentos por cliente (se necessário)
    @GetMapping("/cliente/{clienteId}")
    public ResponseEntity<List<Agendamento>> getAgendamentosByCliente(@PathVariable Long clienteId) {
        try {
            List<Agendamento> agendamentos = agendamentoService.buscarAgendamentosPorCliente(clienteId);
            return ResponseEntity.ok(agendamentos);
        } catch (ResourceNotFoundException e) {
            return ResponseEntity.notFound().build();
        }
    }

    // Exemplo de endpoint para buscar agendamentos por barbeiro (se necessário)
    @GetMapping("/barbeiro/{barbeiroId}")
    public ResponseEntity<List<Agendamento>> getAgendamentosByBarbeiro(@PathVariable Long barbeiroId) {
        try {
            List<Agendamento> agendamentos = agendamentoService.buscarAgendamentosPorBarbeiro(barbeiroId);
            return ResponseEntity.ok(agendamentos);
        } catch (ResourceNotFoundException e) {
            return ResponseEntity.notFound().build();
        }
    }
}
