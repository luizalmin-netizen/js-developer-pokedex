## Estrutura do Prompt para a IA

Para o Copiloto de Vendas, o prompt para a IA deve ser construído de forma a guiar o modelo sobre seu papel e fornecer todo o contexto necessário para gerar respostas úteis e persuasivas. Os elementos chave serão:

1.  **Persona da IA**: Definir o papel do assistente de vendas.
2.  **Contexto da Empresa**: Informações sobre o negócio (`negocio.md`).
3.  **Informações de Produtos**: Detalhes sobre os produtos (`produtos.md`).
4.  **FAQ (Perguntas Frequentes)**: Respostas padronizadas (`faq.md`).
5.  **Estratégias de Objeção**: Como lidar com objeções comuns (`objecoes.md`).
6.  **Mensagem do Cliente**: A entrada real do cliente que a IA precisa responder.

Primeiro, vamos carregar o conteúdo da nossa base de conhecimento para que possamos incluí-lo no prompt.

```python
import os

# Assuming knowledge_dir is already defined and files exist from previous steps
knowledge_dir = 'knowledge'

def load_knowledge_file(filename):
    """Loads content from a specific markdown file in the knowledge directory."""
    file_path = os.path.join(knowledge_dir, filename)
    if os.path.exists(file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            return f.read()
    else:
        print(f"Warning: Knowledge file '{filename}' not found at '{file_path}'.")
        return ""

# Load all knowledge base contents
produtos_kb = load_knowledge_file('produtos.md')
faq_kb = load_knowledge_file('faq.md')
objecoes_kb = load_knowledge_file('objecoes.md')
negocio_kb = load_knowledge_file('negocio.md')

def create_sales_copilot_prompt(customer_message):
    """Constructs the full AI prompt for the sales copilot."""
    ai_persona = "Você é um assistente de vendas prestativo e persuasivo para a empresa Vendas Rápidas, especializada em Refrigerante FYS."

    prompt = f"""
{ai_persona}

## Contexto da Empresa:
{negocio_kb}

## Informações de Produtos:
{produtos_kb}

## FAQ (Perguntas Frequentes):
{faq_kb}

## Estratégias para Lidar com Objeções:
{objecoes_kb}

## Mensagem do Cliente:
{customer_message}

Com base nas informações acima, gere uma resposta concisa, persuasiva e amigável para o cliente, focando em lidar com objeções e promover os produtos FYS. Se a mensagem for uma objeção, use as estratégias fornecidas. Mantenha a resposta breve e direcionada para continuar a conversa de vendas.
Resposta:
"""
    return prompt

print("Knowledge base content loaded.")
print("Function `create_sales_copilot_prompt` defined.")
```

## Exemplo de Interação: Simulação de Resposta da IA para uma Objeção

Agora que temos a função `create_sales_copilot_prompt` e a base de conhecimento carregada, podemos simular uma interação.

Vamos usar a objeção de exemplo `"Está muito caro! Vocês não tem um desconto?"` e ver como o prompt seria construído e qual seria uma resposta esperada da IA.

```python
# Cliente: Mensagem do cliente com uma objeção
customer_objection_message = "Está muito caro! Vocês não tem um desconto?"

# Gerar o prompt completo para a IA
full_ai_prompt = create_sales_copilot_prompt(customer_objection_message)

# --- Simulação da Resposta da IA (sem um modelo real) ---
# Em um cenário real, você passaria `full_ai_prompt` para um modelo de linguagem (como Gemini, ChatGPT/Azure OpenAI).
# Aqui, vamos simular uma resposta que o modelo geraria, baseada nas 'Estratégias para Lidar com Objeções' de objecoes.md
# e no 'Contexto da Empresa'/'Informações de Produtos' da base de conhecimento.

simulated_ai_response = (
    "Entendo sua preocupação com o preço! O Refrigerante FYS realmente entrega uma experiência premium para o público jovem, com um sabor único e a qualidade de uma marca moderna. "
    "Neste momento, não temos descontos diretos, mas posso te assegurar que o custo-benefício é excelente pela durabilidade e satisfação que oferece. "
    "Que tal eu te contar um pouco mais sobre os benefícios exclusivos do FYS, ou posso te mostrar alguma outra opção que se encaixe melhor no seu orçamento sem perder a qualidade? 😊"
)

print(f"### Mensagem do Cliente:\n{customer_objection_message}\n")
print(f"### Prompt Completo Enviado para a IA (Resumido para visualização):\n{full_ai_prompt[:500]}... (conteúdo completo do KB)\n") # Exibe apenas os primeiros 500 caracteres do prompt
print(f"### Resposta Sugerida pela IA:\n{simulated_ai_response}")
```

Este exemplo demonstra como a IA, munida do prompt completo e da base de conhecimento, poderia formular uma resposta que aborda a objeção, reforça o valor do produto e tenta direcionar a conversa para o próximo passo da venda.

## Integração com APIs de Modelos de Linguagem (OpenAI e Gemini)

Para integrar o `full_ai_prompt` com APIs de modelos de linguagem como OpenAI ou Gemini, você precisará:
1.  Instalar as bibliotecas Python correspondentes.
2.  Obter chaves de API e armazená-las de forma segura (recomenda-se usar o gerenciador de segredos do Colab).
3.  Fazer a chamada da API com o prompt gerado.

### Exemplo com OpenAI API

```python
# Instale a biblioteca OpenAI se ainda não estiver instalada
!pip install openai

import openai
from google.colab import userdata

# Configure sua chave de API OpenAI
# Guarde sua chave no secrets manager do Colab com o nome 'OPENAI_API_KEY'
openai.api_key = userdata.get('OPENAI_API_KEY')

# Certifique-se de ter um modelo disponível (ex: gpt-3.5-turbo)
openai_model = "gpt-3.5-turbo"

try:
    response_openai = openai.chat.completions.create(
        model=openai_model,
        messages=[
            {"role": "system", "content": "Você é um assistente de vendas prestativo e persuasivo."},
            {"role": "user", "content": full_ai_prompt}
        ],
        temperature=0.7,
        max_tokens=200
    )
    print("### Resposta do OpenAI:\n")
    print(response_openai.choices[0].message.content)
except Exception as e:
    print(f"Erro ao chamar a API OpenAI: {e}")
    print("Por favor, verifique se sua chave de API OpenAI está configurada corretamente e se o modelo está disponível.")
```

### Exemplo com Gemini API

```python
# Instale a biblioteca Google Generative AI se ainda não estiver instalada
!pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata

# Configure sua chave de API Gemini
# Guarde sua chave no secrets manager do Colab com o nome 'GOOGLE_API_KEY'
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# Inicialize o modelo Gemini (você pode escolher entre 'gemini-pro', 'gemini-1.5-flash', etc.)
gemini_model = genai.GenerativeModel('gemini-1.5-flash')

try:
    response_gemini = gemini_model.generate_content(full_ai_prompt)
    print("### Resposta do Gemini:\n")
    print(response_gemini.text)
except Exception as e:
    print(f"Erro ao chamar a API Gemini: {e}")
    print("Por favor, verifique se sua chave de API Gemini está configurada corretamente e se o modelo está disponível.")
```
