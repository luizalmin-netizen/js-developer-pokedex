```python
# Define the Kaggle dataset path
kaggle_dataset_path = 'YOUR_KAGGLE_DATASET_OWNER/YOUR_KAGGLE_DATASET_NAME'

# Create a directory to store the dataset
!mkdir -p data

# Download the dataset (this downloads all files in the dataset to the 'data' directory)
!kaggle datasets download -d {kaggle_dataset_path} -p data

# Unzip the downloaded file(s) if necessary (most Kaggle datasets are zipped)
# This command assumes there's a single zip file in the 'data' directory.
# You might need to adjust this based on the specific dataset's file structure.
import os
import zipfile

# List files in the 'data' directory to find the zip file
print("Files in data directory:", os.listdir('data'))

# Find the zip file (assuming it's the only zip file, or you know its name)
zip_files = [f for f in os.listdir('data') if f.endswith('.zip')]

if zip_files:
    for zip_file in zip_files:
        with zipfile.ZipFile(os.path.join('data', zip_file), 'r') as zip_ref:
            zip_ref.extractall('data')
        print(f"Extracted {zip_file} to data/")
else:
    print("No zip files found in the data directory.")
```
